```python
import pandas as pd
```

```python
X = pd.read_csv("train.csv")
```

```python
Y = X.pop("Survived")
X.Age.fillna(X.Age.mean(),inplace=True)#use age's mean to fill the missing data
```

```python
X.describe()
```

```python
X.head()
```

```python
X.drop(["Name", "Ticket", "PassengerId"],axis=1, inplace=True)#Name ticket passengeId did not influence
```

```python
def clean_cabin(x):
    try:
        return x[0]
    except TypeError:
        return "none"

X["Cabin"] = X.Cabin.apply(clean_cabin)#deal with cabin, because cabin is not a number.
```

```python
dummies = pd.get_dummies(X["Sex"],prefix="Sex")
X = pd.concat([X, dummies], axis=1)
X.drop("Sex",axis=1,inplace=True)

X["Cabin"].fillna("Missing",inplace=True)
dummies = pd.get_dummies(X["Cabin"],prefix="Cabin")
X = pd.concat([X, dummies], axis=1)
X.drop("Cabin", axis=1, inplace=True)

X["Embarked"].fillna("Missing",inplace=True)
dummies = pd.get_dummies(X["Embarked"],prefix="Embarked")
X = pd.concat([X, dummies], axis=1)
X.drop("Embarked", axis=1, inplace=True)

# what we do here is to make the value to be number, we can not deal with other form except number.
```

```python
X.describe()
```

```python
from sklearn.linear_model import LogisticRegression
#Assumed you have, X (predictor) and Y (target) for training data set and x_test(predictor) of test_dataset
# Create logistic regression object
model_log = LogisticRegression()
 
# Train the model using the training sets and check score
model_log.fit(X, Y)
model_log.score(X, Y)
```

```python
X_test = pd.read_csv("test.csv")
X_test.Age.fillna(X_test.Age.mean(),inplace=True)#use age's mean to fill the missing data
X_test.Fare.fillna(X_test.Fare.mean(),inplace=True)#use age's mean to fill the missing data
X_test.drop(["Name", "Ticket", "PassengerId"],axis=1, inplace=True)#Name ticket passengeId did not influence

X_test["Cabin"] = X_test.Cabin.apply(clean_cabin)#deal with cabin, because cabin is not a number.

dummies = pd.get_dummies(X_test["Sex"],prefix="Sex")
X_test = pd.concat([X_test, dummies], axis=1)
X_test.drop("Sex",axis=1,inplace=True)

X_test["Cabin"].fillna("Missing",inplace=True)
dummies = pd.get_dummies(X_test["Cabin"],prefix="Cabin")
X_test = pd.concat([X_test, dummies], axis=1)
X_test.drop("Cabin", axis=1, inplace=True)

X_test["Embarked"].fillna("Missing",inplace=True)
dummies = pd.get_dummies(X_test["Embarked"],prefix="Embarked")
X_test = pd.concat([X_test, dummies], axis=1)
X_test.drop("Embarked", axis=1, inplace=True)
```
